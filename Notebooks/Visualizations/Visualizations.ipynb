```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

# file for visualizations of perceived motor competence

# LOCOMOTOR SKILLS
#  1. Rennen, 3. Galloperen, 5. Hinkelen, 7. Loopsprong, 9. Voorw. Springen, 11. Zijw. Gallop 

# OBJECT CONTROL SKILLS
#  2. Bal gooien, 4. Bal vangen, 6. Bal rollen, 8. Bal schoppen, 10. Knuppel, 12. Stuiten bal  
```

```python
df = pd.read_csv('../Data-Files/T0_data.csv')
#df['Zip-code'].fillna('')
print(df.keys())

```

```python
mq_category_general_list = []
mq_general_list = []  

for index, rows in df.iterrows():
    if not math.isnan(rows['MQ catogory']):
        mq_category_general_list.append(int(rows['MQ catogory']))

    if not math.isnan(rows['MQ']):
        mq_general_list.append(rows['MQ'])    


```

```python
locomotive_dict = {'running':[], 'galop': [], 'hop':[], 'run jump':[], 'forward jump': [], 'side galop':[] }
for index, rows in df.iterrows():
    if not math.isnan(rows['MQ catogory']):
        if isinstance(rows['1. Rennen'], str):
            if rows['1. Rennen'].isnumeric():
                locomotive_dict['running'].append(rows['1. Rennen'])
        if isinstance(rows['3. Galloperen'], str):
            if rows['3. Galloperen'].isnumeric():
                locomotive['galop'].append(rows['3. Galloperen'])
```

```python
running_mq_category_dict = {1: [], 2: [], 3: [], 4: []}

mq_category_list = []
for index, rows in df.iterrows():
    if not math.isnan(rows['MQ catogory']):
        if isinstance(rows['1. Rennen'], str):
            if rows['1. Rennen'].isnumeric():
                mq_category_list.append(rows['MQ catogory'])
                if int(rows['1. Rennen']) == 1:
                    running_mq_category_dict[1].append(rows['MQ catogory'])

                elif int(rows['1. Rennen']) == 2:
                    running_mq_category_dict[2].append(rows['MQ catogory'])

                elif int(rows['1. Rennen']) == 3:
                    running_mq_category_dict[3].append(rows['MQ catogory'])

                elif int(rows['1. Rennen']) == 4:
                    running_mq_category_dict[4].append(rows['MQ catogory'])        

            
```

```python
running_dict = {1 : [], 2: [], 3: [], 4: [] }
run_list = []
run_mq_list = []
for index, rows in df.iterrows():
    if isinstance(rows['1. Rennen'], str):
        if rows['1. Rennen'].isnumeric():
            run_list.append(rows['1. Rennen'])

    if isinstance(rows['1. Rennen'], str):
        if not math.isnan(rows['MQ']) and rows['1. Rennen'].isnumeric():
            run_mq_list.append(rows['MQ'])
            if int(rows['1. Rennen']) == 1:
                running_dict[1].append(rows['MQ'])

            elif int(rows['1. Rennen']) == 2:
                running_dict[2].append(rows['MQ'])

            elif int(rows['1. Rennen']) == 3:
                running_dict[3].append(rows['MQ'])

            elif int(rows['1. Rennen']) == 4:
                running_dict[4].append(rows['MQ'])        
```

```python
galop_mq_dict = {1: [], 2:[], 3:[], 4:[]}
galop_mq_category_dict = {1: [], 2:[], 3:[], 4:[]}
galop_list = []
galop_mq_list = []

for index, rows in df.iterrows():
    if isinstance(rows['3. Galloperen'], str):
        if rows['3. Galloperen'].isnumeric():
            galop_list.append(rows['3. Galloperen'])
            if not math.isnan(rows['MQ']):
                galop_mq_list.append(rows['MQ'])
                if int(rows['3. Galloperen']) == 1:
                    galop_mq_dict[1].append(rows['MQ'])

                elif int(rows['3. Galloperen']) == 2:
                    galop_mq_dict[2].append(rows['MQ'])

                elif int(rows['3. Galloperen']) == 3:
                    galop_mq_dict[3].append(rows['MQ'])

                elif int(rows['3. Galloperen']) == 4:
                    galop_mq_dict[4].append(rows['MQ'])

            if not math.isnan(rows['MQ catogory']):
                
                if int(rows['3. Galloperen']) == 1:
                    galop_mq_category_dict[1].append(rows['MQ catogory'])

                elif int(rows['3. Galloperen']) == 2:
                    galop_mq_category_dict[2].append(rows['MQ catogory'])

                elif int(rows['3. Galloperen']) == 3:
                    galop_mq_category_dict[3].append(rows['MQ catogory'])

                elif int(rows['3. Galloperen']) == 4:
                    galop_mq_category_dict[4].append(rows['MQ catogory'])
```

```python
### RUNNING AND MEAN MQ BAR PLOT ###


fig, ax = plt.subplots()

for key, value in running_dict.items():
    x = key
    y = np.mean(value)
    ax.bar(x,y)

ax.set_xlabel('Rennen')
ax.set_ylabel('MQ')
plt.ylim((0, 150))
plt.grid(True, axis='y', alpha=0.25)
plt.title('RUNNING AND MEAN MQ')#
plt.show()

```

```python
### RUNNING - COUNT HISTOGRAM PLOT ###


fig, ax = plt.subplots()
ax.hist(run_list, bins=[0, 1, 2, 3, 4])

ax.set_xlabel('Running') 
plt.title('PERCEIVED RUNNING - COUNT')
plt.grid(True, axis='both', alpha=0.25)
plt.show()   
```

```python
### MQ DISTRIBUTION RUNNING HISTOGRAM ###


fig, ax = plt.subplots()
ax.hist(run_mq_list, label='Running')

ax.set_xlabel('MQ')
ax.set_ylabel('# of observations')

plt.title('MQ DISTRIBUTION RUNNING')
plt.ylim((0, 500))
plt.grid(True, axis='both', alpha=0.25)
plt.show()
```

```python
### RUNNING - MEAN MQ CATEGORY BAR PLOT ###


fig, ax = plt.subplots()

for key, value in running_mq_category_dict.items():
    x = key
    y = np.mean(value)
    ax.bar(x,y)

ax.set_xlabel('Running')
ax.set_ylabel('MQ Category')
plt.ylim((0,5))
plt.grid(True, axis='y', alpha=0.25)
plt.title('RUNNING - MEAN MQ CATEGORY')
plt.show()
```

```python
### MQ CATEGORY DISTRIBUTION RUNNING HISTOGRAM ###


fig, ax = plt.subplots()

ax.hist(mq_category_list, bins=5)

ax.set_xlabel('MQ Category')
ax.set_ylabel('# of observations')
plt.title('MQ CATEGORY DISTRIBUTION RUNNING')
plt.grid(True, axis='both', alpha=0.25)
plt.show()
```

```python
### GALOP - MEAN MQ BAR PLOT ###


fig, ax = plt.subplots()

for k, v in galop_mq_dict.items():
    x = k
    y = np.mean(v)
    ax.bar(x,y)

ax.set_xlabel('Galop')
ax.set_ylabel('MQ')
plt.ylim((0, 150))
plt.title('GALOP - MEAN MQ')
plt.grid(True, axis='y', alpha=0.25)
plt.show()
```

```python
### PERCEIVED GALOP DISTRIBUTION ###

fig, ax = plt.subplots()
ax.hist(galop_list, bins= 4)

ax.set_xlabel('Galop') 
ax.set_ylabel('# of observations')
plt.title('PERCEIVED GALOP DISTRIBUTION')
plt.grid(True, axis='both', alpha=0.25)
plt.show()   
```
