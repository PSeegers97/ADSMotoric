```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

# file for visualizations of perceived motor competence

# LOCOMOTOR SKILLS
#  1. Rennen, 3. Galloperen, 5. Hinkelen, 7. Loopsprong, 9. Voorw. Springen, 11. Zijw. Gallop 

# OBJECT CONTROL SKILLS
#  2. Bal gooien, 4. Bal vangen, 6. Bal rollen, 8. Bal schoppen, 10. Knuppel, 12. Stuiten bal  
```

```python
df = pd.read_csv('../Output-Files/t0_wo_nan.csv')
#df['Zip-code'].fillna('')
print(df.keys())
print(df.info())
```

```python
def HistoPlotsMQ(df):
    col_dict = {1 : [], 2: [], 3: [], 4: [], 5:[] }
    pcm_dict = {1: [], 2: [], 3: [], 4:[]}
    
    
    for col in df.columns:
        col_list = []
        
        unique_values_list = df[col].unique()

        if 1 in unique_values_list and 2 in unique_values_list and 3 in unique_values_list and 4 in unique_values_list and 5 in unique_values_list and len(unique_values_list) ==5:
            print('\033[1m' + col)
            for ind, rows in df.iterrows():
               
                if not col == 'MQ' or not col == 'MQ category':
                    col_list.append(int(rows[col]))
                    if int(rows[col]) == 1:
                        col_dict[1].append(rows['MQ'])
                    elif int(rows[col]) == 2:
                        col_dict[2].append(rows['MQ'])
                    elif int(rows[col]) == 3:
                        col_dict[3].append(rows['MQ']) 
                    elif int(rows[col]) == 4:
                        col_dict[4].append(rows['MQ']) 
                    elif int(rows[col]) == 5:
                        col_dict[5].append(rows['MQ'])
                    
                    
            fig, ax = plt.subplots()

            for key, value in col_dict.items():
                x = key
                y = np.mean(value)
                ax.bar(x,y)
            plt.title(col)
            ax.set_xlabel(col)
            ax.set_ylabel('MQ score mean')    
            plt.show()    

            fig, ax = plt.subplots()
            ax.hist(col_list) 
            ax.set_ylabel('# of observations')
            ax.set_xlabel(col) 
            plt.title(col)
            plt.show()
        
        if 1 in unique_values_list and 2 in unique_values_list and 3 in unique_values_list and 4 in unique_values_list and len(unique_values_list) ==4:
            print('\033[1m' + col) 
            for ind, rows in df.iterrows():
                
                if not col == 'MQ' or not col == 'MQ category':
                    col_list.append(int(rows[col]))
                    if int(rows[col]) == 1:
                        pcm_dict[1].append(rows['MQ'])
                    elif int(rows[col]) == 2:
                        pcm_dict[2].append(rows['MQ'])
                    elif int(rows[col]) == 3:
                        pcm_dict[3].append(rows['MQ']) 
                    elif int(rows[col]) == 4:
                        pcm_dict[4].append(rows['MQ']) 
                    elif int(rows[col]) == 5:
                        pcm_dict[5].append(rows['MQ'])
                    
                    
            fig, ax = plt.subplots()

            for key, value in pcm_dict.items():
                x = key
                y = np.mean(value)
                ax.bar(x,y)
            plt.title(col)
            ax.set_xlabel(col)
            ax.set_ylabel('MQ score mean')    
            plt.show()    

            fig, ax = plt.subplots()
            ax.hist(col_list) 
            ax.set_ylabel('# of observations')
            ax.set_xlabel(col) 
            plt.title(col)
            plt.show()
        
    
```
